```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
%matplotlib inline
```

Рассмотрим экспоненциальное распределение

```python
expon_rv = expon()
sample = expon_rv.rvs(1000)
```

Эмпирическая функция распределения для полученной выборки:

```python
x = np.linspace(0,10,1000)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf, label='theoretical CDF')
plt.show()
```

Построим гистаграмму.
Построим график функции плотности вероятности:

```python
pdf = expon_rv.pdf(x)
plt.hist(sample, normed=True)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Сгенируем 1000 выборок объема n = 5, 10, 50

```python
means_5 = [np.mean(expon_rv.rvs(5)) for i in range(1000)]
stds_5 = [np.std(expon_rv.rvs(5)) for i in range(1000)]
means_10 = [np.mean(expon_rv.rvs(10)) for i in range(1000)]
stds_10 = [np.std(expon_rv.rvs(10)) for i in range(1000)]
means_50 = [np.mean(expon_rv.rvs(50)) for i in range(1000)]
stds_50 = [np.std(expon_rv.rvs(50)) for i in range(1000)]
```

```python
plt.figure(figsize=(20,10))
plt.hist(means_5, bins = 50, normed = True, label = "5",alpha=0.7)
plt.hist(means_10, bins = 50, normed = True, label = '10',alpha=0.7)
plt.hist(means_50, bins = 50, normed = True, label="50", alpha=0.7)
plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

```python
import scipy.stats as sts
x = np.linspace(0,3,1000)
```

```python
plt.figure(figsize=(20,10))
norm_rv_5=sts.norm(loc=np.mean(means_5), scale=np.std(means_5))
pdf_5 = norm_rv_5.pdf(x)
norm_rv_10=sts.norm(loc=np.mean(means_10), scale=np.std(means_10))
pdf_10 = norm_rv_10.pdf(x)
norm_rv_50=sts.norm(loc=np.mean(means_50), scale=np.std(means_50))
pdf_50 = norm_rv_50.pdf(x)
plt.plot(x, pdf_5, label = "pdf_5")
plt.plot(x, pdf_10, label = "pdf_10")
plt.plot(x, pdf_50,label = "pdf_50")
plt.hist(means_5, bins = 50, normed = True, label = "5",alpha=0.7)
plt.hist(means_10, bins = 50, normed = True, label = '10',alpha=0.7)
plt.hist(means_50, bins = 50, normed = True, label="50", alpha=0.7)
plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.show()
```

Как видим на графике с увеличением выборки распределение средних стремится к нормальному закону.
