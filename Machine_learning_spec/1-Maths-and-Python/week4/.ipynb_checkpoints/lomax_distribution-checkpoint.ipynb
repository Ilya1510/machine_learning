## Центральная предельная теорема на примере распределения lomax

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Выберите ваше любимое непрерывное распределение (чем меньше оно будет похоже на нормальное, тем интереснее; попробуйте выбрать какое-нибудь распределение из тех, что мы не обсуждали в курсе). Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины (чтобы величины были в одном масштабе, не забудьте выставить у гистограммы значение параметра normed=true).

```python
# Функция, которая строит нужную гистограмму и теоретическую плотность распределения lomax(c)
# c - параметр распределения, right_board_plot - правая граница графика по X
# bins_count - число столбцов в гистограмме, num - число точек на theorethical cdf
# size - размер выборки для построения гистограммы
def draw_lomax_pdf(c = 2.5, right_board_plot = 5, bins_count = 18, num = 1000, size = 1000):
    lomax_rv = sts.lomax(c)
    sample = lomax_rv.rvs(size=size, random_state = 1)
    x = np.linspace(0, right_board_plot, num = num)

    plt.ylabel('fraction of samples')
    plt.xlabel('$x$')
    plt.hist(sample, normed=True, bins=x[::num // bins_count], label='hist of samples')
    pdf = lomax_rv.pdf(x)
    plt.plot(x, pdf, label='theoretical PDF')
    plt.legend(loc='upper right')
```

```python
draw_lomax_pdf()
```

Ваша задача — оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок. Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних. Используя информацию о среднем и дисперсии исходного распределения (её можно без труда найти в википедии), посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Обратите внимание: для подсчёта значений этих параметров нужно использовать именно теоретические среднее и дисперсию вашей случайной величины, а не их выборочные оценки. Поверх каждой гистограммы нарисуйте плотность соответствующего нормального распределения (будьте внимательны с параметрами функции, она принимает на вход не дисперсию, а стандартное отклонение).

Из википедии: $mean = \frac{1}{c - 1} = \frac{1}{1.5} = \frac{2}{3}$ $varience = \frac{c}{(c - 1) ^ 2 * (c - 2)} = \frac{2.5}{1.5 ^ 2 * 0.5} = \frac{20}{9}$ 


```python
# Функция, которая строит гистограмму выборочных средних и плотность нормального распределения,
# которое по ЦПТ должно приближать гистограмму.
# c - параметр распределения, n - кол-во объектов в выборке(из скольких будет браться выборочное среднее)
# size - кол-во выборок(на скольких точках будет строится гистограмма), right_board_plot - правая граница плота по X
# bins_count - число корзин в гистограмме, num - число точек на графике нормального распределения 
def draw_hist_sample_mean_distribution(c = 2.5, n = 5, size = 10000, 
                                       right_board_plot = 5, bins_count = 18, num = 1000):
    lomax_rv = sts.lomax(c)
    mean_samples = []
    for i in range(size):
        mean_samples.append(np.mean(lomax_rv.rvs(n)))

    plt.ylabel('fraction mean samples')
    plt.xlabel('$valuemeansamples$')
    x = np.linspace(0, right_board_plot, num = num)
    plt.hist(mean_samples, normed=True, bins=x[::num // bins_count], label='hist of mean samples')
    
    # N(EX, DX / n)
    mean = 1 / (c - 1)
    varience = c / (c - 1) / (c - 1) / (c - 2) / n
    norm_rv_cpt = sts.norm(mean, np.sqrt(varience)) 
    pdf = norm_rv_cpt.pdf(x)
    plt.title('norm pdf and mean samples hist n = ' + str(n))
    plt.plot(x, pdf, label='norm PDF')
    plt.legend(loc='upper right')
    plt.show()
```

```python
draw_hist_sample_mean_distribution(n=3, bins_count=30, right_board_plot=4)
```

```python
draw_hist_sample_mean_distribution(n=10, bins_count=36, right_board_plot=3)
```

```python
draw_hist_sample_mean_distribution(n=50, bins_count=50, right_board_plot=3)
```

```python
draw_hist_sample_mean_distribution(n=500, bins_count=200, right_board_plot=2)
```

```python
draw_hist_sample_mean_distribution(n=10000, bins_count=600, right_board_plot=1)
```

## Результаты:
Видно, что чем больше число n, тем более точно приближает нормальное распределение из ЦПТ искомое распределение. Которое, кстати, довольно сильно скошено и даже при n = 50 гистограмма не очень напоминает нормальное распределение. Но вот при n = 10000, усомнится в этом невозможно.
